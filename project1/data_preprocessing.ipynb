# Food waste (bubble)

```python
import pandas as pd

# population
population = pd.read_csv("population.csv")

# gdp
gdp = pd.read_csv("gdp.csv")

# region 
region = pd.read_csv('region.csv')
```

```python
## extract population and gdp details (year 2010)
population_2010 = population.iloc[:, [0, -12]] 
gdp_2010 = gdp.iloc[:, [0, -12]]

# extract region details for each country
region = region.iloc[:, [0, 2]]
```

```python
# inner join based on country name (population and gdp)
merge_df = pd.merge(population_2010, gdp_2010, on='Country Name', how='inner')
merge_df.columns = ['Country', 'population', 'gdp']

# inner join based on country name (region)
merge_df = pd.merge(merge_df, region, on='Country', how='inner')

# rename columns
merge_df.columns = ['country', 'population', 'gdp', 'region']
```

```python
# food waste
food_waste_2010 = pd.read_csv("food_waste.csv")
```

```python
food_waste_2010
```

```python
# extract required columns
food_waste_2010 = food_waste_2010[['country', 'loss_percentage']]

food_waste_2010.loc[food_waste_2010['country'].str.contains("United States of America"), 'country'] = 'United States'
```

```python
# group by country and aggregate using sum operation
food_waste = food_waste_2010.groupby('country').agg('sum')
```

```python
# merge food waste with population, gdp and region
final_df = pd.merge(merge_df, food_waste, on='country', how='inner')
```

```python
final_df.to_csv('food_waste_donut.csv', encoding='utf-8', index=False)
```

# Food waste (radial bar)

```python
food_waste = pd.read_csv('food_waste_2010.csv')
food_waste
```

```python
food_waste['food_supply_stage'] = food_waste['food_supply_stage'].replace(['Retail'],'Retail & Wholesale')
food_waste['food_supply_stage'] = food_waste['food_supply_stage'].replace(['Wholesale'],'Retail & Wholesale')
```

```python
food_waste = food_waste.drop(food_waste[food_waste['food_supply_stage'].str.contains('Farm')].sample(frac=.4).index)
```

```python
# group by food supply stage and aggregate using sum operation
food_waste = food_waste.groupby('food_supply_stage')['loss_percentage'].agg('sum')

#print(food_waste.groupby('food_supply_stage')['loss_percentage'].agg('sum'))
```

```python
food_waste.to_csv('food_waste_radial.csv', encoding='utf-8', index=True)
```

# Food waste (parallel)

```python
import pandas as pd

food_waste_cereal = pd.read_csv('food_waste_cereal.csv')
food_waste_fruit = pd.read_csv('food_waste_fruit.csv')
food_waste_other = pd.read_csv('food_waste_other.csv')
food_waste_crop = pd.read_csv('food_waste_crop.csv')
food_waste_meat = pd.read_csv('food_waste_meat.csv')

# region 
region = pd.read_csv('region.csv')
```

```python
food_waste_cereal['major waste type'] = 'Cereals & Pulses'
food_waste_fruit['major waste type'] = 'Fruits & Vegetables'
food_waste_other['major waste type'] = 'Other'
food_waste_crop['major waste type'] = 'Roots, Tubers & Oil-Bearing Crops'
food_waste_meat['major waste type'] = 'Meat & Animals Products'
```

```python
food_waste = pd.concat([food_waste_cereal, food_waste_fruit, food_waste_other, food_waste_crop, food_waste_meat]).reset_index(drop=True)
food_waste
```

```python
region = region.iloc[:, [0, 2]]
```

```python
region.columns = ['country', 'Region']
```

```python
# inner join based on country name (population and gdp)
merge_df = pd.merge(food_waste, region, on='country', how='inner')
merge_df
```

```python
merge_df.to_csv('food_waste_parallel.csv', encoding='utf-8', index=False)
```

# New Food Waste (bubble)

```python
import pandas as pd

# population
population = pd.read_csv("population.csv")

# gdp
gdp = pd.read_csv("gdp.csv")

# region 
region = pd.read_csv('region.csv')
```

```python
population = population.reset_index()
population = pd.melt(population, id_vars='Country Name', value_vars=list(map(str, list(range(1980, 2011)))), var_name='year', value_name='population')
```

```python
gdp = gdp.reset_index()
gdp = pd.melt(gdp, id_vars='Country Name', value_vars=list(map(str, list(range(1980, 2011)))), var_name='year', value_name='gdp')
```

```python
population
```

```python
gdp
```

```python
merge_df = pd.concat([population.reset_index(drop=True), gdp], axis=1)
merge_df
```

```python
merge_df = merge_df[['Country Name', 'year', 'population', 'gdp']]
merge_df
```

```python
merge_df = merge_df.loc[:, ~merge_df.columns.duplicated()]
merge_df
```

```python
merge_df.columns = ['Country', 'year', 'population', 'gdp']

# extract region details for each country
region = region.iloc[:, [0, 2]]

# inner join based on country name (region)
merge_df = pd.merge(merge_df, region, on='Country', how='inner')
```

```python
merge_df
```

```python
merge_df.columns = ['country', 'year', 'population', 'gdp', 'region']

# food waste
food_waste = pd.read_csv("food_waste_1965_2010.csv")
```

```python
food_waste
```

```python
# extract required columns
food_waste = food_waste[['country', 'year', 'loss_percentage']]

food_waste.loc[food_waste['country'].str.contains("United States of America"), 'country'] = 'United States'

food_waste
```

```python
merge_df['year'] = merge_df['year'].astype(int)
```

```python
final_df = pd.merge(merge_df, food_waste, on=['country', 'year'], how='inner')
final_df
```

```python
final_df = final_df.groupby(['country','year','population','gdp','region']).agg('sum')
```

```python
final_df.to_csv('food_waste_bubble_new.csv', encoding='utf-8', index=True)
```
