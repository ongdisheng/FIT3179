```python
import pandas as pd

child_death = pd.read_csv('child-deaths.csv')
child_mortality = pd.read_csv('child-mortality.csv')
```

```python
child_death_2020 = child_death[child_death['Year'] == 2020]
child_mortality_2020 = child_mortality[child_mortality['Year'] == 2020]
```

```python
child_death_2020
```

```python
child_mortality_2020
```

```python
merge_df = pd.merge(child_death_2020, child_mortality_2020, on=['Entity', 'Code', 'Year'])
```

```python
merge_df
```

```python
merge_df['Entity'].unique()
```

```python
merge_df = merge_df[['Entity', 'Number of under-five deaths', 'Mortality rate, under-5 (per 1,000 live births)']]
merge_df.columns = ['Country', 'Number_of_child_deaths', 'Child_mortality_rate_per_1000_live_births']
```

```python
merge_df
```

```python
merge_df.to_csv('child-mortality-2020.csv')
```
