# Bubble plot

```python
import pandas as pd

bubble = pd.read_csv('child-mortality-gdp-per-capita.csv')
```

```python
bubble
```

```python
bubble = bubble[['Entity', 'Year', 'Child mortality (Select Gapminder, v10) (2017)', 'GDP per capita', 'Population (historical estimates)']]
```

```python
bubble.columns = ['country', 'year', 'child_mortality', 'gdp_per_capita', 'population']
```

```python
continent = pd.read_csv('continent.csv')
```

```python
continent
```

```python
merge = pd.merge(bubble, continent, on=['country'])
```

```python
merge
```

```python
final = merge[(merge['year'] >= 2000) & (merge['year'] <= 2016)]
```

```python
final.to_csv('bubble.csv')
```
