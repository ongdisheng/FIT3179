# Bubble plot

```python
import pandas as pd

child_mortality = pd.read_csv('child-mortality-gdp-per-capita.csv')
child_death = pd.read_csv('child-deaths-igme-data.csv')
continent = pd.read_csv('continent.csv')
```

```python
child_mortality
```

```python
child_death
```

```python
continent
```

```python
# extract columns
child_mortality = child_mortality[['Entity', 'Year', 'Child mortality (Select Gapminder, v10) (2017)', 'GDP per capita', 'Population (historical estimates)']]
child_death = child_death[['Entity', 'Year', 'Number of under-five deaths']]
```

```python
# change column names
child_mortality.columns = ['country', 'year', 'child_mortality_rate', 'gdp_per_capita', 'population']
child_death.columns = ['country', 'year', 'child_mortality']
```

```python
final = pd.merge(child_mortality, child_death, on=['country', 'year'])
final = pd.merge(final, continent, on=['country'])
final
```

```python
final = final[(final['year'] >= 1990) & (final['year'] <= 2016)]
final
```

```python
final.to_csv('bubble_plot.csv')
```

# Causes of death (bump chart)

```python
import pandas as pd

cause = pd.read_csv('causes-of-death-in-children.csv')
```

```python
cause
```

```python
cause.columns = ['country', 'code', 'year', 'Malaria', 'HIV/AIDS', 'Meningitis', 'Nutritional deficiencies', 'Other neonatal disorders', 'Whooping cough', 'Lower respiratory infections', 'Congenital birth defects', 'Measles', 'Neonatal sepsis & infections', 'Birth asphyxia & trauma', 'Drowning', 'Tuberculosis', 'Preterm birth', 'Diarrheal diseases', 'Neoplasms', 'Syphilis']
```

```python
cause = cause.melt(id_vars=['country', 'code', 'year'], var_name='cause', value_name='deaths')
cause
```

```python
cause = cause[['year', 'cause', 'deaths']]
cause
```

```python
final = cause.groupby(['year', 'cause']).aggregate('sum').reset_index()
```

```python
final
```

```python
final = final[(final['year'] >= 1990) & (final['year'] <= 2016)]
final
```

```python
final.to_csv('bump_chart_cause.csv')
```

# Slope chart

```python
import pandas as pd

slope1 = pd.read_csv('child-deaths.csv')
slope2 = pd.read_csv('child-deaths-by-life-stage.csv')
slope3 = pd.read_csv('infant-deaths.csv')
```

```python
slope1
```

```python
slope1 = slope1[slope1['Indicator'] == 'Deaths age 5 to 9']
slope1
```

```python
slope1 = slope1[['Geographic area', 'TIME_PERIOD', 'OBS_VALUE']]
slope1
```

```python
slope1.columns = ['country', 'year', 'deaths']
slope1
```

```python
slope1 = slope1[(slope1['year'] == '1990-06') | (slope1['year'] == '2016-06')]
slope1
```

```python
slope1['year'] = slope1['year'].replace({'1990-06': 1990, '2016-06': 2016})
slope1['life_stage'] = 'Year 5-9'
slope1
```

```python
slope2
```

```python
slope3
```

```python
slope2 = slope2[['Entity', 'Year', 'Child deaths age 1 to 4']]
slope3 = slope3[['Entity', 'Year', 'Deaths - All causes - Sex: Both - Age: 28-364 days (Number)', 'Deaths - All causes - Sex: Both - Age: 7-27 days (Number)', 'Deaths - All causes - Sex: Both - Age: 0-6 days (Number)']]
```

```python
slope2.columns = ['country', 'year', 'Year 1-4']
slope3.columns = ['country', 'year', 'Day 28-364', 'Day 7-27', 'Day 0-6']
```

```python
slope2
```

```python
slope3
```

```python
final = pd.merge(slope2, slope3, on=['country', 'year'])
final
```

```python
final = final.melt(id_vars=['country', 'year'], var_name='life_stage', value_name='deaths')
final
```

```python
final = final[(final['year'] == 1990) | (final['year'] == 2016)]
final
```

```python
slope1 = slope1[['country', 'year', 'life_stage', 'deaths']]
slope1
```

```python
final = pd.concat([slope1, final])
final
```

```python
final = final[['year', 'life_stage', 'deaths']]
final
```

```python
final = final.groupby(['year', 'life_stage']).aggregate('mean').reset_index()
final
```

```python
final.to_csv('slope_chart.csv')
```

# Stacked area chart

```python
import pandas as pd

stacked = pd.read_csv('child-deaths-igme-data.csv')
```

```python
stacked
```

```python
stacked = stacked[(stacked['Year'] >= 1990) & (stacked['Year'] <= 2016)]
stacked
```

```python
stacked.columns = ['country', 'code', 'year', 'deaths']
stacked
```

```python
stacked = stacked[['country', 'year', 'deaths']]
```

```python
continent = pd.read_csv('continent.csv')
continent
```

```python
final = pd.merge(stacked, continent, on='country', how='left')
final
```

```python
final[final['continent'].isna()]['country'].unique()
```

```python
final.loc[final.country == 'Congo', 'continent'] = 'Africa'
final.loc[final.country == "Cote d'Ivoire", 'continent'] = 'Africa'
final.loc[final.country == "Czechia", 'continent'] = 'Europe'
final.loc[final.country == "Democratic Republic of Congo", 'continent'] = 'Africa'
final.loc[final.country == "Micronesia (country)", 'continent'] = 'Oceania'
final.loc[final.country == "Timor", 'continent'] = 'Asia'
```

```python
final[final['continent'].isna()]['country'].unique()
```

```python
final
```

```python
final = final[['year', 'deaths', 'continent']]
final
```

```python
final = final.groupby(['year', 'continent']).aggregate('sum').reset_index()
final
```

```python
final.to_csv('stacked_area_chart.csv')
```
