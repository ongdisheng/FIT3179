```python
import pandas as pd

child_death = pd.read_csv('child-deaths.csv')
pop = pd.read_csv('population.csv')
```

```python
child_death_2020 = child_death[child_death['Year'] == 2020]
pop_2020 = pop[pop['Year'] == 2020]
```

```python
child_death_2020
```

```python
pop_2020
```

```python
merge_df = pd.merge(child_death_2020, pop_2020, on='Entity')
```

```python
merge_df
```

```python
merge_df['Entity'].unique()
```

```python
merge_df = merge_df[['Entity', 'Number of under-five deaths', 'Population (historical estimates)']]
merge_df.columns = ['Country', 'Number of under-five deaths', 'Population']
```

```python
merge_df
```

```python
merge_df.to_csv('child-mortality-2020.csv')
```
